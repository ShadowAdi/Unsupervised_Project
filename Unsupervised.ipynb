```python
df=df.sample(frac=1)
```

```python
cols=["Fresh","Milk","Grocery","Frozen","Detergents_Paper","Delicassen"]
```

### Standard Scaling

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
df[cols]=ss.fit_transform(df[cols])
```

```python
df.sample(1)
```

### Applying PCA

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca_components = pca.fit_transform(df)
```

```python
pca_components
```

```python
from sklearn.cluster import KMeans


pca_df = pd.DataFrame(data=pca_components, columns=['Principal Component 1', 'Principal Component 2'])
```

```python
pca_df.sort_values(by="Principal Component 2",ascending=False)
```

```python
sns.scatterplot(x=pca_df["Principal Component 1"],y=pca_df["Principal Component 2"])
```

```python
pca_df=pca_df[pca_df["Principal Component 1"]<= 8.0]
```

```python
pca_df=pca_df[pca_df["Principal Component 2"]<= 6.0]
```

```python
sns.scatterplot(x=pca_df["Principal Component 1"],y=pca_df["Principal Component 2"])
```

### KMeans Clustering

```python
wcss=[]
from sklearn.cluster import KMeans
for i in range(1,11):
    kmeans = KMeans(n_clusters=i,init="k-means++")
    labels_train_kmean = kmeans.fit_predict(pca_df)
    wcss.append(kmeans.inertia_)
```

```python
wcss
```

```python
plt.plot(range(1,11),wcss)
plt.title("WCSS Vs K-No")
plt.xlabel("No. Of Components")
plt.ylabel("WCSS")
plt.show()
```

```python
silhote_scores=[]
from sklearn.metrics import silhouette_score
for i in range(2,11):
    kmeans = KMeans(n_clusters=i, init="k-means++")
    labels_train_kmeans = kmeans.fit_predict(pca_df)

    silhouette_avg = silhouette_score(pca_df, labels_train_kmeans)
    silhote_scores.append(silhouette_avg)

```

```python
plt.plot(range(2,11),silhote_scores)
plt.title("Silhoute Scores Vs K-No")
plt.xlabel("No. Of K")
plt.ylabel("Silhoute Scores")
plt.show()
```

```python
kmeans = KMeans(n_clusters=2,init="k-means++")
labels_kmean = kmeans.fit_predict(pca_df)
labels_kmean
```

```python
silhouette_avg = silhouette_score(pca_df, labels_kmean)
silhouette_avg
```

```python
pca_df
```

```python
sns.scatterplot(x=pca_df["Principal Component 1"],y=pca_df["Principal Component 2"],hue=labels_kmean)
```

### DBSCAN

```python
from sklearn.cluster import DBSCAN
silhoute_avg_dbscan=[]
for i in range(1,11):
        dbscan = DBSCAN(eps=0.5, min_samples=i)
        labels_train_dbscan = dbscan.fit_predict(pca_df)
        silhouette_avg = silhouette_score(pca_df, labels_train_dbscan)
        silhoute_avg_dbscan.append(silhouette_avg)

        
```

```python
plt.plot(range(1,11),silhoute_avg_dbscan)
plt.title("Silhoute Scores Vs DBScan")
plt.xlabel("Min Samples")
plt.ylabel("DBScan")
plt.show()
```

```python
dbscan = DBSCAN(eps=0.5, min_samples=4)
labels_train_dbscan = dbscan.fit_predict(pca_df)
silhouette_avg = silhouette_score(pca_df, labels_train_dbscan)
print(f"Silhouette Score is: {silhouette_avg}")
```

```python
sns.scatterplot(x=pca_df["Principal Component 1"],y=pca_df["Principal Component 2"],hue=labels_train_dbscan)
```

### **For This Case We Are Considering DBScan to  Be Our Best Model**
### Because The Score Come is `0.61` for `DBScan` and `0.52` for `KMeansClustering`.
